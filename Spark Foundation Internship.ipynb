### *What are total Sales and Profits of the company by segment?*

```python
Sales_and_Profits = df.groupby("Segment").sum().iloc[:,[8,-1]].sum()
round(Sales_and_Profits,2) # Rounding Numericala entries to 2 decimal places 
```

### *Top 10 States by Sales and Profits*

```python
Top_10_Sales = df.groupby("State").Sales.sum().nlargest(n =10)
Top_10_Profits = df.groupby("State").Profit.sum().nlargest(n =10)
```

```python
Top_10_Sales 
```

### Top 10 States by Sales:

### Top 10 States by Profit:

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn')
ax = Top_10_Sales.plot(kind ='bar', figsize =(14,8), fontsize =14)
ax.set_xlabel("States", fontsize =13)
ax.set_ylabel("Total Sales",fontsize =13)
ax.set_title("Top 10 States by Sales",fontsize =16)
plt.show()  
```

```python
plt.style.use('seaborn')
ax = Top_10_Profits.plot(kind ='bar', figsize =(14,8), fontsize =14)
ax.set_xlabel("States", fontsize =13)
ax.set_ylabel("Total Profits",fontsize =13)
ax.set_title("Top 10 States by Profits",fontsize =16)
plt.show()   
```

#### California and New York are far ahead of their competitor states.

```python
# Pair_plot
financial = df.loc[:,['Sales','Quantity','Discount','Profit']]
sns.pairplot(financial)  
```

#### In above we see that there is some relation between sales and profit and also there is some relation between Discount and Profit. Now To see what exact relation between those entities we plot the heat_map. so we get more clearity

```python
correlation=financial.corr()
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True) 
```

#### From Above map we infer that, 
#### 1) Sales and Profits are positively correlated
#### 2) Discount and Profits are Negatively correlated

##### Hence we consider these cases and proceed further

## Case-1 When Discount is 0


```python
data=df[df['Discount']==0]
sns.relplot(x='Sales',y='Profit',data=data) 
```

#### Hence we can say that there is positive relation between Profit and sales when Discount is 0 so Now we will plot heat_map to get correlaton. 

```python
numerical_features = data.select_dtypes(include='number') 
correlation = numerical_features.corr()
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True)  
```

#### Hence, we see that There is strong correlation between sales and Profit i.e 0.92 

```python
#Rel_plot with respect category
sns.relplot(x='Sales',y='Profit',hue='Category',data=data) 
```

#### Furniture category is not making profits whereas Technology category is making huge profit and sales at the same time when discount = 0 

## Case-2 When Discount is Not 0

```python
data1=df[df['Discount']!=0]
sns.relplot(x='Sales',y='Profit',hue='Discount',data=data1) 
```

#### In above graph we can see that as percentage of Discount increases the sales also increses but profit decreases
#### Now let's check how it can effect on different sectors of business

```python
sns.relplot(x='Sales',y='Profit',hue='Category',data=data1) 
```

```python
numerical_features = data1.select_dtypes(include='number') 
correlation=numerical_features.corr()
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True)
```

#### Here, we can se that there is a negative correlation between sales and profit when discount!=0 unlike when discount is 0

```python
sns.set(rc={'figure.figsize':(18,10)})
sns.barplot(x = 'Sub-Category' , y = 'Profit' , data = data1) 
```

#### Here we see that copiers has highest Profit and Machines has highest loss

```python
sns.set(rc={'figure.figsize':(18,10)})
sns.barplot(x = 'Sub-Category' , y = 'Sales' , data = data1) 
```

##### Here we see that copiers has highest sale and Machines has second highest sales

##### In above two graph we can see that 'Machines' has second highest sale but due to large discount, it is in loss and in second graph we can see that sales in 'Fasteners','labels'and 'Art' category are so weak, so we have to concentrate on these sub-category businesses

#### Discounts are triggering losses 

## Distrubution of Profits across diffrent regions by Segments 

```python
plt.figure(figsize = (12,4))
sns.set(font_scale=1, palette= "viridis")
sns.barplot(data = df , x = "Region",y = "Profit" ,hue = "Segment")
plt.show() 
```

#### Overall each Segment is profitable. 

## Profit distribution across Region by category 

```python
plt.figure(figsize = (12,4))
sns.set(font_scale=1, palette= "viridis")
sns.barplot(data = df , x = "Region",y = "Profit" ,hue = "Category")
plt.show()
```

#### "Furniture" Category is the only loss making sector and that to only in Central Region

##  Investigating losses in Furniture category in the Central region

```python
# Grouping Data by Region and only slicing Data for Central Region from whole Data Set
gb_Central = df[df['Region'] == 'Central']
gb_Central 
```

```python
# Investing Further in cenral Region 
plt.figure(figsize = (12,4))
sns.set(font_scale=1.5, palette= "viridis")
sns.barplot(data = gb_Central, x = "Category",y = "Profit" ,hue = "Ship Mode")
plt.title("Investigation of central region: Profit making(by Ship Mode)")
plt.show() 
```

#### Losses are inccured in Furniture Cateory irrespective to ship mode in Central Region

## Investigating individual performance by states in the central region

```python
gb_Category_Furniture = df[(df['Region'] == 'Central') & (df['Category'] == 'Furniture')]            
gb_Category_Furniture 
```

```python
plt.figure(figsize = (12,8))
sns.set(font_scale=1, palette= "viridis")
sns.barplot(data = gb_Category_Furniture , x = "State",y = "Profit" ,hue = "Sub-Category")
plt.title("Investigation of Central Region Furniture Category: Profit Analysis(by Sub Category)", fontsize = 20)
plt.show() 
```

```python
plt.figure(figsize = (12,8))
sns.set(font_scale=1, palette= "viridis")
sns.barplot(data = gb_Category_Furniture , x = "State",y = "Profit" ,hue = "Segment")
plt.title("Investigation of Central Region Furniture Category: Profit Analysis(by Segment)", fontsize = 20)
plt.show()
```

### Texas and Illiois are only two states contributing to all the losses in Furniture category in the Central Region
### Losses in Tables Sub Category is significanlty high.

## *So, what is it they are doing diffrently?* 

```python
plt.figure(figsize = (12,8))
sns.set(font_scale=1, palette= "viridis")
sns.barplot(data = gb_Category_Furniture , x = "State",y = "Discount" ,hue = "Sub-Category")
plt.title("Discounts provided by each state", fontsize = 20)
plt.show() 
```

## Texas and Illinois are only states providing discounts in the whole central region this justifies high positive correlation between postal codes and discounts.
## Also, these discounts are very high!
## *1. 60% on Furnishings*
## *2. 30% on Bookcases and Chairs*
## *3. 50% disount on Tables in Illinois and 30% in Texas*
## Are these discounts driving sales up?.... NO!

```python
plt.figure(figsize = (10,10))
sns.set(font_scale=1.5)
sns.lmplot(data = gb_Category_Furniture , x = "Discount", y ="Sales", aspect = 1, height = 8, col ="Sub-Category", col_wrap= 2)
plt.show() 
```

#### Sales tend to do down when discounts go up in 3 out of 4 "Subcategories" and also in overall Furniture Category Sales in the Central Region.
